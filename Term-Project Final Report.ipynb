```python
class Food :
    def __init__(self, name, price, stock) :
        self.Name = name
        self.Price = price
        self.Stock = stock
        self.Chosen = 0
        self.Left = stock
    def LED(self) :
        if (self.Stock <= 0) : 
            self.Led = 1
        else : self.Led = 0

class Money :
    def __init__(self, name, stock) :
        self.Name = name
        self.Stock = stock
    def LED(self) :
        if (self.Stock <= 0) : 
            self.Led = 1
        else : 
            self.Led = 0
```

```python
def ShowFStock(a) :
    a.LED()
    if (a.Led == 1) : print(a.Name + " 재고없음")
```

```python
def ShowMStock(a) :
    a.LED()
    if (a.Led == 1) : print("\t\t\t\t\t\t" + a.Name + "잔돈없음")
    else : print("\t\t\t\t\t\t" + a.Name + str(a.Stock) + "개 남음")
```

```python
def order() :
    #장바구니
    order = [0]
    while(1) :
        order.append(int(input(ham.Name + "(1) " + coke.Name + "(2) 장바구니(3) 결제(4)>> ")))
        if order[-1] == 1 :
            ham.Chosen += 1 
            ham.Stock -= 1 
            if(ham.Stock < 0) : 
                ham.Stock += 1
                ham.Chosen -= 1
                print(ham.Name + " 재고 x")
            print("--------------------------------------------------------------------")
        elif order[-1] == 2 :
            coke.Chosen += 1
            coke.Stock -= 1
            if(coke.Stock < 0) : 
                coke.Stock += 1
                coke.Chosen -= 1
                print(coke.Name + " 재고 x")
            print("--------------------------------------------------------------------")
        elif order[-1] == 3 : 
            basket(order)
            print("--------------------------------------------------------------------")
        elif order[-1] == 4 :
            if (ham.Chosen == 0) and (coke.Chosen == 0) : 
                print("▷음식을 골라주세요")
                continue
            else :
                if ham.Chosen > 0 :
                    print(ham.Name + " " + str(ham.Chosen) + "개 주문하셨습니다.")
                if coke.Chosen > 0 :
                    print(coke.Name + " " + str(coke.Chosen) + "개 주문하셨습니다.")
                break
        else : continue
```

```python
def basket(order) :  
    if ham.Chosen > 0 :
        print("장바구니에 " + ham.Name + " " + str(ham.Chosen) + "개 담겼습니다.")
    if coke.Chosen > 0 :
        print("장바구니에 " + coke.Name + " " + str(coke.Chosen) + "개 담겼습니다.")
    total = ham.Chosen * 1000 + coke.Chosen * 500
    print("현재 " + str(total) + "원입니다.")
```

```python
def cash(total) :
  #투입금액
    money = 0
    while(total > money) :
        now_input = input2charge(int(input("현금을 넣어주세요 >> ")))
        money =  now_input + money
        print("--------------------------------------------------------------------")
        print("총 : " + str(total) + "원 투입금액 : " + str(money) + "원")
    money = money - total
    print("거스름돈 : " + str(money) + "원")
    print("--------------------------------------------------------------------")
    cashChange(money)
```

```python
def input2charge(money) :
    if money / 10000 == 1 :
        print("10000원을 넣었습니다.")
        return 10000
        m_10000.Stock += 1
    elif money / 5000 == 1 :
        print("5000원을 넣었습니다.")
        m_5000.Stock += 1
        return 5000
    elif money / 1000 == 1 :
        print("1000원을 넣었습니다.")
        m_1000.Stock += 1
        return 1000
    elif money / 500 == 1 :
        print("500원을 넣었습니다.")
        m_500.Stock += 1
        return 500
    elif money / 100 == 1 :
        print("100원을 넣었습니다.")
        m_100.Stock += 1
        return 100
    else :
        print("지폐나 동전을 하나씩 넣어주세요.")
        return 0
```

```python
def cashChange(money) :
    total = money
    c_10000 = money // 10000
    if (c_10000 > m_10000.Stock) :
        c_10000 = m_10000.Stock
        money -= 10000 * m_10000.Stock
    else : money -= 10000 * c_10000
    m_10000.Stock -= c_10000

    c_5000 = money // 5000
    if (c_5000 > m_5000.Stock) :
        c_5000 = m_5000.Stock
        money -= 5000 * m_5000.Stock
    else : money -= 5000 * c_5000
    m_5000.Stock -= c_5000

    c_1000 = money // 1000
    if (c_1000 > m_1000.Stock) :
        c_1000 = m_1000.Stock
        money -= 1000 * m_1000.Stock
    else : money -= 1000 * c_1000
    m_1000.Stock -= c_1000

    c_500 = money // 500
    if (c_500 > m_500.Stock) :
        c_500 = m_500.Stock
        money -= 500 * m_500.Stock
    else : money -= 500 * c_500
    m_500.Stock -= c_500

    c_100 = money // 100
    if (c_100 > m_100.Stock) :
        c_100 = m_100.Stock
        money -= 100 * m_100.Stock
    else : money -= 100 * c_100
    m_100.Stock -= c_100

    if (c_5000 * 5000 + c_1000 * 1000 + c_500 * 500 + c_100 * 100 < total) : 
        print("잔돈이 부족합니다. 반환된 돈을 받으세요")
        m_5000.Stock += c_5000
        m_1000.Stock += c_1000
        m_500.Stock += c_500
        m_100.Stock += c_100
    else :
        print("5000원 : " + str(c_5000) + "개")
        print("1000원 : " + str(c_1000) + "개")
        print("500원 : " + str(c_500) + "개")
        print("100원 : " + str(c_100) + "개")
```

```python
def card(total) :
    while(1) : 
        money = int(input("카드를 넣어주세요(1) >> "))
        print("카드를 인식중입니다.")
        if (money == 1) : 
            print(str(total) + "원이 결제되었습니다.")
            break
        else : print("카드 인식 오류")
    print("--------------------------------------------------------------------")
```

```python
import random

#main()

#음식 재고 입력
ham = Food("햄버거", 2000, 100)
coke = Food("콜라", 1000, 100)

#잔돈 재고 입력
m_10000 = Money("\t\t10000원 지폐", 1)
m_5000 = Money("\t\t 5000원 지폐", 1)
m_1000 = Money("\t\t 1000원 지폐", 6)
m_500 = Money("\t\t 500원 동전", 10)
m_100 = Money("\t\t 100원 동전", 20)

#대기번호 계산 시 필요한 랜덤 대기번호
a = random.randrange(100, 111)
b = random.randrange(900, 911)

#시작
while(1) :
    print("\t 𝓦𝓮𝓵𝓬𝓸𝓶𝓮 𝓽𝓸 𝓢𝓸𝓸'𝓼 𝓻𝓮𝓼𝓽𝓪𝓾𝓻𝓪𝓷𝓽")
    print()
    #음식 재고
    ShowFStock(ham)
    ShowFStock(coke)

    #잔돈 확인
    ShowMStock(m_10000)
    ShowMStock(m_5000)
    ShowMStock(m_1000)
    ShowMStock(m_500)
    ShowMStock(m_100)
    print()

    #주문
    order()
  
    #총액
    total = ham.Chosen * 1000 + coke.Chosen * 500
    print("총 " + str(total) + "원 입니다.")
    print("--------------------------------------------------------------------")

    #매장 식사 or 포장
    
    while(1) :
        place = int(input("매장 식사(1) or 포장(2) >> ")) 
        if place != 1 and place != 2:
            print("▷다시 선택해주세요")
        else : break
    print("--------------------------------------------------------------------")
  
    #결제수단 & 영수증 발급여부
    while(1) :
        payment = int(input("현금(1) or 카드(2) >> "))  
        print("--------------------------------------------------------------------")
        if payment == 1 : 
            cash(total)
            while(1) :
                rec = int(input("현금영수증을 발급할까요? 네(1) 아니요(2) >> "))
                if rec == 1 :
                    receipt()
                    break
                elif rec == 2 :
                    break
                else :
                    print("▷다시 선택해주세요")
                    continue
            break
        elif payment == 2 : 
            card(total)
            while(1) :
                rec = int(input("영수증을 발급할까요? 네(1) 아니요(2) >> "))
                if rec == 1 :
                    receipt()
                    break
                elif rec == 2 :
                    break
                else :
                    print("▷다시 선택해주세요")
                    continue
            break
        else : 
            print("▷다시 선택해주세요")
            continue
  
    #대기표 발급
    if place == 1 :
        print("대기번호 : " + str(a) + "번")
        a += 1
    elif place == 2 :
        print("대기번호 : " + str(b) + "번")
        b += 1
    print("--------------------------------------------------------------------")

    #초기화
    ham.Chosen = 0
    coke.Chosen = 0
    print()
    print()
    print()
```
